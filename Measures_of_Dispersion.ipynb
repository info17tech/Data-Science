NUMPY

Range, Variance, Standard Deviation

```python
import numpy as np
```

```python
speed = [86,87,88,86,87,85,86]
```

```python
min_speed = np.min(speed)
max_speed = np.max(speed)

range_speed = max_speed - min_speed

print("Max:", max_speed)
print("Min:", min_speed)
print("Range:", range_speed)
```

```python
# Default: ddof=0
p_var = np.var(speed)
s_var = np.var(speed, ddof=1)
print("Population Variance:", p_var)
print("Sample Variance:", s_var)
```

```python
p_stdev = np.std(speed)
s_stdev = np.std(speed, ddof=1)
print("Population Standard Deviation:", p_stdev)
print("Sample Standard Deviation:", s_stdev)
```

PANDAS

```python
import pandas as pd
```

```python
score = [32,111,138,28,59,77,97]
```

```python
data = pd.Series(score)
```

```python
min_score = data.min()
max_score = data.max()

range_score = max_score - min_score

print("Max:", max_score)
print("Min:", min_score)
print("Range:", range_score)
```

```python
# Default: ddof=1
p_var = data.var(ddof=0)
s_var = data.var(ddof=1)
print("Population Variance:", p_var)
print("Sample Variance:", s_var)
```

```python
p_stdev = data.std(ddof=0)
s_stdev = data.std(ddof=1)
print("Population Standard Deviation:", p_stdev)
print("Sample Standard Deviation:", s_stdev)
```
