NUMPY

Mean, Median and Mode

```python
import numpy as np
```

```python
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
mean_speed = np.mean(speed)
print("Mean:", mean_speed)
```

```python
median_speed = np.median(speed)
print("Median:", median_speed)
```

```python
from scipy import stats
mode_speed = stats.mode(speed)
print("Mode:", mode_speed.mode)
```

IQR - To identify the outliers

```python
q1 = np.percentile(speed, 25)
q3 = np.percentile(speed, 75)

iqr = q3 - q1

print("Q1:", q1)
print("Q3:", q3)
print("IQR is", iqr)
```

```python
upper_bound = q3 + 1.5 * iqr
lower_bound = q1 - 1.5 * iqr
print("Upper bound:", upper_bound)
print("Lower bound:", lower_bound)
```

```python
speed_array = np.array(speed)
outlier = speed_array[(speed_array < lower_bound) | (speed_array > upper_bound)]
print(outlier, type(outlier))
```

PANDAS

```python
import pandas as pd
```

```python
score = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
series = pd.Series(score)
```

```python
mean_score = series.mean()
print("Mean:", mean_score)
```

```python
median_score = series.median()
print("Median:", median_score)
```

```python
mode_score = series.mode()
print("Mode:", mode_score)
```

IQR

```python
data = pd.Series(score)
```

```python
q1 = data.quantile(0.25)
q3 = data.quantile(0.75)

iqr = q3 - q1

print("Q1:", q1)
print("Q3:", q3)
print("IQR is", iqr)
```

```python
upper = q3 + 1.5 * iqr
lower = q1 - 1.5 * iqr
print("Upper bound:", upper)
print("Lower bound:", lower)
```

```python
outlier = data[(data < lower) | (data > upper)]
print(outlier)
```

```python
clean_data = data[(data >= lower) | (data <= upper)]
print(clean_data)
```
