```python
import numpy as np
```

```python
study_hours = [1,2,3,4,5,6,7,8]
marks = [35,40,50,55,65,70,80,90]

std_x = np.nanstd(study_hours)
std_y = np.nanstd(marks)

print("STD of study hours: ", std_x)
print("STD of marks: ", std_y)
```

Correlation Coefficient

```python
corr_coef = np.corrcoef(study_hours, marks)
print("Correlation: ", corr_coef)
```

```python
mean_x = np.mean(study_hours)
mean_y = np.mean(marks)
print("Mean of study_hours: ", mean_x)
print("Mean of marks: ", mean_y)
```

```python
slope = corr_coef[0, 1] * (std_y / std_x)
print("Slope: ", slope)
```

```python
coeff_regr = np.polyfit(study_hours, marks, 1)
print("Coeff of regression: ", coeff_regr)
```

```python
slope = coeff_regr[0]
print("Slope: ", slope)
```

```python
y_intercept = coeff_regr[1]
print("Y-intercept: ", y_intercept)
```

Best Fit Line -> y = slope * study_hours + y_intercept

Residual calculation

```python
y_predicted = slope * np.array(study_hours) + y_intercept

residuals = np.array(marks) - y_predicted

print("Predicted Marks: ", y_predicted)
print("\nResiduals: ", residuals)
```

```python
squared_errors = residuals ** 2
print("Squared Errors: ", squared_errors)
```

Sum of Squared Errors is the Unexplained Error

```python
SSE = np.sum(squared_errors)
print("SSE: ", SSE)
```

```python
deviation = np.array(marks) - mean_y
print("Deviation: ", deviation)
```

```python
squared_deviation = deviation ** 2
print("Squared Deviation: ", squared_deviation)
```

Total Sum of Squares is the total variation (in Marks)

```python
SST = np.sum(squared_deviation)
print("SST: ", SST)
```

Regression Sum of Squares is explained by study hours

```python
SSR = SST - SSE
print("SSR: ", SSR)
```

Coefficient of determination (R2) is the predictive power

```python
R2 = SSR/SST * 100
print("R2 in % : ", R2)
```
