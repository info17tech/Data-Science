```python
import pandas as pd
```

```python
data = {
    "StudyHours":[2,3,4,5,6,7,8,9,10],
    "Attendance":[60,65,70,75,80,85,90,92,95],
    "Sleep":[6,6,7,7,7,8,8,8,9],
    "Marks":[45,50,58,65,72,80,88,91,96]
}
```

```python
df = pd.DataFrame(data)
df
```

## y=b0+b1x1+b2x2+b3x3
### Marks=b0+b1* study hours+b2* attendance+b3* sleephours

Goal is to find b0, b1, b2, b3

```python
from sklearn.linear_model import LinearRegression

# X is independent variables, Y is a dependant variable
X = df[["StudyHours", "Attendance", "Sleep"]]
Y = df["Marks"]
```

```python
model = LinearRegression()
model.fit(X, Y)

print("Bo-Intercept: ", model.intercept_)
print("Coefficient: ", model.coef_)
```

### Marks = -32.352000000000075 + 1.24 * study_hours + 1.148 * attendance + 0.78533333 * sleep_hours

β₁ = 1.24 (StudyHours)

If study hours increase by 1 hour, marks increase by 1.24 points, holding attendance and sleep constant.

β₂ = 1.148 (Attendance)

If attendance increases by 1%, marks increase by 1.148 points, holding other variables constant.

β₃ = 0.785 (Sleep)

If sleep increases by 1 hour, marks increase by 0.785 points, holding other variables constant.

Predict the marks

```python
# Method 1
StudyHours = 8
Attendance = 90
SleepHours = 8

Marks = -32.352000000000075 + 1.24 * StudyHours + 1.148 * Attendance + 0.78533333 * SleepHours
print("Predicted Marks: ", Marks)
```

```python
# Method 2
new_student = pd.DataFrame(
    {"StudyHours": [8], "Attendance": [90], "Sleep": [8]}
    )
new_student
```

```python
marks_prediction = model.predict(new_student)
print("Predicted Marks: ", marks_prediction)
```

Finding out SSE, SST, SSR for prediction line (R2)

```python
X
```

Calculate the Residuals

```python
df['Predicted_Marks'] = model.predict(X)
df
```

```python
# Find the residual
df['Residual'] = df['Marks'] - df['Predicted_Marks']
df
```

```python
df['Residual_Squared'] = df['Residual'] ** 2
df
```

Calculate SSE=∑e2

```python
SSE = df['Residual_Squared'].sum()
print('SSE:', SSE)
```

Calculate SST

```python
y_mean = df['Marks'].mean()
print('Mean of Y:', y_mean)
```

```python
# y - y_mean
df['y_mean_diff'] = df['Marks'] - y_mean
df
```

```python
df['y_mean_diff_sq'] = df['y_mean_diff'] ** 2
df
```

```python
SST = df['y_mean_diff_sq'].sum()
print('SST:', SST)
```

Calculate SSR

```python
SSR = SST - SSE
print('SSR:', SSR)
```

Calculate R square

```python
R_square = SSR / SST
print('R Square:', R_square)
```

```python
# Find R2 of variation percentage
R2_percentage = R_square * 100
print('R2 percentage:', R2_percentage)
```

Find R^2 using scikit learn

```python
R_square = model.score(X, Y)
print('R Square:', R_square)
```

Adjusted R^2

adjusted_R2 = 1 - (1-R2)*(n-1)/(n-k-1)

R² from sklearn

n = number of observations

k = number of predictors

```python
n = len(Y)      # number of rows
k = X.shape[1]  # number of predictors (columns)
print('Number of observations:', n)
print('Number of predictors:', k)

adjusted_R2=1-(1-R_square)*(n-1)/(n-k-1)
print('Adjusted R_square:',adjusted_R2)
```
